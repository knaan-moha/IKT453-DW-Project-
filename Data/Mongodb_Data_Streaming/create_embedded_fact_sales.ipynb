```python
import pandas as pd
import os
```

```python
# Init dim and fact files directory
dim_courier_status_dir = '../dim_courier_status.csv'
dim_customer_dir = '../dim_customer.csv'
dim_date_dir = '../dim_date.csv'
dim_fulfillment_dir = '../dim_fulfillment.csv'
dim_product_dir = '../dim_product.csv'
dim_promotion_dir = '../dim_promotion.csv'
dim_status_dir = '../dim_status.csv'
fact_dir = '../fact_sales.csv'
```

```python
# Read fact sales file and its corresponding dim tables(files)
df_dim_courier_status = pd.read_csv(dim_courier_status_dir)
df_dim_customer = pd.read_csv(dim_customer_dir)
df_dim_date = pd.read_csv(dim_date_dir)
df_dim_fulfillment = pd.read_csv(dim_fulfillment_dir)
df_dim_product = pd.read_csv(dim_product_dir)
df_dim_promotion = pd.read_csv(dim_promotion_dir)
df_dim_status = pd.read_csv(dim_status_dir)
df_fact_sales = pd.read_csv(fact_dir)
```

### Creating embedded fact_sales to upload in mongoDB

```python
# Merge/join dimension(s) in fact_sales
df_embedded = df_fact_sales \
    .merge(df_dim_date, on="date_id", how="left") \
    .merge(df_dim_customer, on="customer_id", how="left") \
    .merge(df_dim_product, on="product_id", how="left") \
    .merge(df_dim_fulfillment, on="fulfillment_id", how="left") \
    .merge(df_dim_promotion, on="promotion_id", how="left") \
    .merge(df_dim_status, on="status_id", how="left") \
    .merge(df_dim_courier_status, on="courier_status_id", how="left")

df_embedded
```

```python
# Load embedded fact sales to csv file
data_dir = '../../Data'
df_embedded.to_csv(os.path.join(data_dir, 'fact_sales_embedded.csv'), index=False)
```
